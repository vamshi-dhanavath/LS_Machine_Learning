```python
from ultralytics import YOLO

model = YOLO('yolov8.yaml')
model = YOLO('yolov8m.pt')
model = YOLO('yolov8m.yaml').load("yolov8m.pt")
model.train(data="/content/drive/MyDrive/Learner Space/ML/Assignment_2/data.yaml",
            epochs=30,
            imgsz=300,
            lr0=0.06,
            batch=32)
metrics = model.val()
```
